```python
import json_helper as jh
import pandas as pd
import json
import os
import numpy as np
```

```python
def read_json(file_path: str):
    with open(file_path) as f:
        json_input = json.load(f)
    return json_input

def read_all_json_files(JSON_ROOT):
    result = []
    for dirpath, dirname, filenames in os.walk(JSON_ROOT):
        
        for f in filenames:
            if f.endswith('.json'):
                json_content = read_json(os.path.join(JSON_ROOT, f))
                for i in json_content['results']:
                    i['source'] = f
                    result.append(i)
    df_location = pd.DataFrame(result)
    return df_location
```

```python
# display a summary of df_daily_summaries
df_daily_summaries = read_all_json_files('./daily_summaries')
```
